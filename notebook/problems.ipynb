## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

def lanzar_dosdados():
    dado1 = random.randint(1,6)
    dado2 = random.randint(1,6)
    return dado1 + dado2

lanzamientos = 1000
flag = 0
for i in range(lanzamientos):
    res = lanzar_dosdados()
    if res >7 or res % 2 == 0:
        flag +=1

print(f"La probabilidad estimada de que la suma sea par o mayor que 7 es: {flag/lanzamientos}")
```

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

#Definir funciÃ³n coger bola
n_bolas=60
def coger_bola():
    randomnumber=random.randint(0,59)
    return randomnumber
# Variables para las simulaciones
n_simulaciones = 1000
count=0
iguales=0

#Caso 3 blancas y 2 rojas
for i in range(n_simulaciones):
    res = {}
    flagred=0
    flagwhite=0
    for j in range(5):
        res[j] = ball_box[coger_bola()]
        if res[j] == 'White':
            flagwhite+=1
        elif res[j] == 'Red':
            flagred+=1
    if flagwhite == 3 and flagred == 2:
        count+=1
print(f"La probabilidad estimada de que haya 3 blancas y 2 rojas en 1000 simulaciones es: {count/n_simulaciones}")
         
#Caso todas mismo color
for i in range(n_simulaciones):
    res = {}
    flagred=0
    flagwhite=0
    flaggreen=0
    for j in range(5):
        res[j] = ball_box[coger_bola()]
        if res[j] == 'White':
            flagwhite+=1
        elif res[j] == 'Red':
            flagred+=1
        else:
            flaggreen+=1
    if flagwhite == 5 or flagred == 5 or flaggreen == 5:
        iguales+=1
print(f"La probabilidad estimada de que presenten el mismo color en 1000 simulaciones es: {iguales/n_simulaciones}")
```
